# GaRField

A Gaussian Random Field generator that produces multi-resolution realizations while preserving the large-scale Fourier modes of the original field.

## Modules

```python
# Matplotlib
import matplotlib.pyplot as plt

# Numpy
import numpy as np
from numpy.random import default_rng

# Scipy
from scipy.stats import norm

# Sys
import sys
```

```python
# GaRField
sys.path.append("../garfield")
import gaussianrandomfield as gaussrf
```

## GRF

### Parameters

IO directory and file

```python
grf_dir  = "../data/"
grf_file = grf_dir + "grf.npy"
```

ND array shape

```python
grf_shape = np.array((128, 128, 128))
```

Gaussian distribution: mean $\mu$ and variance $\sigma^2$

```python
grf_mu    = 0.0
grf_sigma = 1.0
```

Integer seed for RNG

```python
grf_seed = 1
```

### Data

Init RNG with seed

```python
rng = default_rng(grf_seed)
```

Generate new field and save

```python
grf = gaussrf.generate(grf_mu, grf_sigma, grf_shape, rng)
print("Save", grf_file)
np.save(grf_file, grf)
```

Check statistics

```python
print("Mean :", np.mean(grf))
print("Std  :", np.std( grf))
```

Plot distribution

```python
xmax = grf_mu + 4*grf_sigma
xmin = grf_mu - 4*grf_sigma
x = np.linspace(xmin, xmax, 100)
y = norm.pdf(x, grf_mu, grf_sigma)

plt.figure()
plt.hist(grf.ravel(), bins=100, density=True)
plt.plot(x, y, 'k')
plt.xlim(xmin, xmax)
plt.xlabel('x')
plt.ylabel('PDF')
plt.show()
```
